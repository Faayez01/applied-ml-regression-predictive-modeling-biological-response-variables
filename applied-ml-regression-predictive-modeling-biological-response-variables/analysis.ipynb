```
Fertilizer <- read.csv(url("https://whitlockschluter3e.zoology.ubc.ca/Data/chapter17/chap17q20GrasslandNutrientsPlantSpecies.csv"))
Fertilizer
```

```
ggplot(Fertilizer, aes(nutrients, species)) + 
    geom_point(size = 3, col = "firebrick") + 
    labs(x = "Number of nutrients added", y = "Number of plant species") + 
    theme_classic()
```

1: A. The scatter plot of the data is shown above. The number of nutrients added should be the explanatory variable (X) whereas the number of plant species should be the response variable (Y).

```
FRegression <- lm(species ~ nutrients, data=Fertilizer)
FRegression
summary(FRegression)
```

B. The rate of change in the number of plant species supported per nutrient type added is -3.339 and the standard error for the estimate is 1.098

```
ggplot(Fertilizer, aes(x=nutrients, 
   y=species)) +
   geom_point() +
   theme_minimal() +
   xlab("Number of nutrients added") +
   ylab("Number of plant species") +    
   geom_smooth(method = lm)
```

C. The updated scatter plot with the least squares regression line is shown above. The fraction of the variation in the
number of plant species “explained” by the number of nutrients added is 0.536 as given by the R-squared value.

D. We reject the null hypothesis of no treatment effect on the number of plant species as the Regression P-value of 0.01607 < 0.05 i.e. statistically significant. Moreover, the slope i.e. rate of change in the number of plant species supported per nutrient type added is non-zero.


```
Mutations <- read.csv(url("https://whitlockschluter3e.zoology.ubc.ca/Data/chapter17/chap17q35FatherAgeMutations.csv"))
Mutations
```

```
ggplot(Mutations, aes(AgeOfFather, numberOfNewMutations)) + 
    geom_point(size = 3, col = "firebrick") + 
    labs(x = "Age of Father", y = "Number of New Mutations") + 
    theme_classic()
```

```
ggplot(Mutations, aes(x=AgeOfFather, 
   y=numberOfNewMutations)) +
   geom_point() +
   theme_minimal() +
   xlab("Age of Father") +
   ylab("Number of New Mutations") +    
   geom_smooth(method = lm)
```

2: A. The relationship between number of new mutations (Y) and father’s age (X) with the regression line is shown above.

```
MRegression <- lm(numberOfNewMutations ~ AgeOfFather, data=Mutations)
MRegression
summary(MRegression)
```

B. Based on these data, the number of new mutations increase by 1.583/year with father’s age as given by the slope or rate of change. The standard error for the estimate is 0.2543

```
(1.5828*18) + 15.2090 # prediction: Age = 18; Y = bX + a
```

```
(1.5828*36) + 15.2090 # prediction: Age = 36
```

C. The predicted mean number of new mutations from fathers 36 years of age is 72.19 compared to 43.7 for 18 years of age as per the Regression Equation. It is important to note that the predicted values are slightly offset from the actual reported in the data i.e. 70 and 41 due to the residual effect for the least squares regression line of best fit.

```
Mutationsaov <- aov(numberOfNewMutations ~ AgeOfFather, data = Mutations)
summary(Mutationsaov)
```

D. We reject the null hypothesis of no relationship between father’s age and number of new mutations, based on a P-value of 5.584e-06 < 0.05 i.e. statistically significant as indicated in the ANOVA table. Moreover, the slope i.e. rate of change is non-zero.

E. The fraction of the variation among fathers in the number of new mutations explained by father’s age is 0.671 as given by the R-squared value.
